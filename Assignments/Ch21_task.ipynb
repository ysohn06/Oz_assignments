[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n):
    #0부터 n까지의 숫자들 준비하기
    for i in range(n + 1):
        #그 숫자들이 홀수인지 확인하기
        if i % 2 == 1:
            #홀수가 맞으면 생성하기
            yield i

#for문 사용해서 1부터 10까지 홀수만 생성하기
for i in odd_numbers(10):
    print(i)
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max():
    #사용자로부터 3개의 숫자 입력받아서 리스트로 만들기(공백)
    numbers = list(map(int, input().split()))
    #리스트를 튜플로 변환하기
    numbers_tuple = tuple(numbers)
    #최소값과 최대값 찾아서 출력하기
    print(min(numbers_tuple))
    print(max(numbers_tuple))

#함수 실행하기
min_max()
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(n):
    #주어진 숫자가 짝수인지 확인하기
    if n % 2 == 0:
        #짝수이면 True 반환하기
        return True
    #위 조건이 맞지 않다면 False 반환하기
    return False

#함수 실행하기
print(is_even(33))
print(is_even(22))
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    #n이 1이거나 0이되면 1을 돌려주고 멈추게하기
    if n <= 1:
        return 1

    #n과 그보다 하나 작은 숫자(n-1)의 팩토리얼을 곱하기
    return n * factorial(n-1)

#함수 싱행하기
print(factorial(20))
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_length(words):
    #리스트 안에 있는 단어들 꺼내기
    for word in words:
        #len함수 이용해서 길이 출력하기
        print(len(word))

#함수 실행하기
list = ["pizza", "burger", "burrito"]
print_length(list)
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(str1, str2):
    #첫 번째 문자열 길이가 두번째보다 길거나 같다는 가정하기
    if len(str1) >= len(str2):
        return str1
    #그렇지 않다면
    else:
        return str2

#함수 실행하기
print(longer_string("hungry", "tired"))
```
